```python
demographic_df = pd.read_csv('racial_demographics_raw.csv')

white_pop = []
asian_pop = []
hisp_pop = []
black_pop = []

for index, row in demographic_df.iterrows():
    white_males = row['WA_MALE']
    white_females = row['WA_FEMALE']
    black_males = row['BA_MALE']
    black_females = row['BA_FEMALE']
    asian_males = row['AA_MALE']
    asian_females = row['AA_FEMALE']
    hisp_male = row['H_MALE']
    hisp_female = row['H_FEMALE']
    white = white_males + white_females
    black = black_males + black_males
    asian = asian_males + asian_males
    hisp = hisp_male + hisp_female
    white_pop.append(white)
    asian_pop.append(asian)
    black_pop.append(black)
    hisp_pop.append(hisp)

demographic_df['white_pop'] = white_pop
demographic_df['asian_pop'] = asian_pop
demographic_df['black_pop'] = black_pop
demographic_df['hisp_pop'] = hisp_pop

demographic_df.to_csv('racial_demographics.csv', index=False)
```

```python
csv_name = 'confirmed_deaths_10_9_2021.csv'
master_df = pd.read_csv(csv_name)
demographic_df = pd.read_csv('racial_demographics.csv')
white_pop = []
asian_pop = []
hisp_pop = []
black_pop = []
for index, row in master_df.iterrows():
    fips = row['FIPS']
    try:
        white = int(demographic_df.loc[demographic_df['fips'] == fips]['white_pop'])
        asian = int(demographic_df.loc[demographic_df['fips'] == fips]['asian_pop'])
        black = int(demographic_df.loc[demographic_df['fips'] == fips]['black_pop'])
        hisp = int(demographic_df.loc[demographic_df['fips'] == fips]['hisp_pop'])

        white_pop.append(white)
        asian_pop.append(asian)
        hisp_pop.append(hisp)
        black_pop.append(black)
    except Exception:
        white_pop.append('')
        asian_pop.append('')
        hisp_pop.append('')
        black_pop.append('')



master_df['white_pop'] = white_pop
master_df['asian_pop'] = asian_pop
master_df['hisp_pop'] = hisp_pop
master_df['black_pop'] = black_pop

master_df = master_df.loc[master_df['white_pop'] != '']

master_df.to_csv(csv_name, index=False)
```

```python
#Cell by Julie -- adding poverty % to the dataset

#Bringing in poverty by fips
orig_poverty_df = pd.read_csv('PovertyEstimates.csv',encoding="ISO-8859-1", skipinitialspace=True)

#Taking in only identifying factors and the total percent of pop in poverty for 2019
data = [orig_poverty_df['FIPStxt'],orig_poverty_df['Stabr'],orig_poverty_df['Area_name'],orig_poverty_df['PCTPOVALL_2019']]
cols=['FIPS','StateAbr','Admin2','POVERTY_PERCENT']
poverty_df = pd.DataFrame(data, cols)
poverty_df = poverty_df.transpose()
#display(poverty_df)

#we need to get rid of the "county" part of the county name
poverty_df['Admin2'] = poverty_df['Admin2'].str.rstrip('County')
#doing it again to strip the trailing whitespace that is only sometimes there
poverty_df['Admin2'] = poverty_df['Admin2'].str.strip() 

#we need the full state name to match the master dataset, not just the abbreviaton
state = poverty_df['StateAbr'].map(abbrev_to_us_state)
poverty_df['Province_State']=state
#display(poverty_df)
#display(master_df)

# #debugging -- to help me get a better column match 
# #the lengths of county names appeared to be diffeerent b/c of trailing whitespace
# pv = poverty_df.iloc[3]
# ms = master_df.iloc[1]
# pv_s = pv['Admin2']
# ms_s = ms['Admin2']
# print(pv_s,ms_s)
# print(len(pv_s),len(ms_s)) #says pv_s is 1char more (8)
# print(pv_s[7]) #last element is space
#end debugging stuff
#success -- currently matching 3008 columns

#map fips and poverty percent column into the master_df by state/county pair
master_df = pd.merge(poverty_df, master_df, on=['Province_State','Admin2'])
display(master_df)

#current shape --> 3008 rows x 16 cols
```

```python
original_df = pd.read_csv('time_series_covid19_deaths_US.csv')
fips_df = original_df[['FIPS', 'Combined_Key']]

def fill_fips_code_for_df(csv_name):
    df_to_fill = pd.read_csv(csv_name)
    fips_codes = []
    print(df_to_fill.shape)
    for index, row in df_to_fill.iterrows():
        county = row['Combined_Key']
        try:
            fips = fips_df.loc[fips_df['Combined_Key'] == county]['FIPS']
            fips_codes.append(int(fips))
        except ValueError:
            fips_codes.append(0)
    print(len(fips_codes))
    df_to_fill['FIPS'] = fips_codes
    df_to_fill = df_to_fill.loc[df_to_fill['FIPS'] != 0]
    df_to_fill.to_csv(csv_name, index=False)
    print(df_to_fill)
fill_fips_code_for_df('confirmed_deaths_10_9_2021.csv')
```

```python
original_df = pd.read_csv('time_series_covid19_deaths_US.csv')
population_df = original_df[['Population', 'FIPS']]

def fill_population_for_df(csv_name):
    df_to_fill = pd.read_csv(csv_name)
    populations = []
    print(df_to_fill.shape)
    for index, row in df_to_fill.iterrows():
        county_fips = row['FIPS']
        try:
            population = population_df.loc[population_df['FIPS'] == county_fips]['Population']
            populations.append(int(population))
        except ValueError:
            populations.append(0)
    print(len(populations))
    df_to_fill['Population'] = populations
    df_to_fill = df_to_fill.loc[df_to_fill['Population'] != 0]
    df_to_fill.to_csv(csv_name, index=False)
    print(df_to_fill)
fill_population_for_df('confirmed_deaths_10_9_2021.csv')
```

```python
population_density_df = pd.read_csv('Population-Density By County.csv')
population_density_df = population_density_df[['fips', 'Density per square mile of land area']]

def fill_population_density_for_df(csv_name):
    df_to_fill = pd.read_csv(csv_name)
    population_densities = []
    # print(df_to_fill.shape)
    for index, row in df_to_fill.iterrows():
        county_fips = row['FIPS']
        try:
            pop_density_row = population_density_df.loc[population_density_df['fips'] == county_fips]
            population_density = pop_density_row['Density per square mile of land area']
            population_densities.append(float(population_density))
        except ValueError:
            population_densities.append(0)
        except Exception:
            population_densities.append(0)
    df_to_fill['pop per sq mi'] = population_densities
    df_to_fill = df_to_fill.loc[df_to_fill['pop per sq mi'] != 0]
    df_to_fill.to_csv(csv_name, index=False)
fill_population_density_for_df('confirmed_deaths_5_1_2020.csv')
```

```python
vote_df = pd.read_csv('countypres_2000-2020_2.csv')
vote_df = vote_df[['county_fips', 'party', 'candidatevotes']]

def fill_vote_for_df(csv_name):
    df_to_fill = pd.read_csv(csv_name)
    democrat_votes = []
    republican_votes = []
    for index, row in df_to_fill.iterrows():
        county_fips = row['FIPS']
        try:
            county_vote_rows = vote_df.loc[vote_df['county_fips'] == county_fips]
            democrat_vote_row = county_vote_rows.loc[county_vote_rows['party'] == 'DEMOCRAT']
            republican_vote_row = county_vote_rows.loc[county_vote_rows['party'] == 'REPUBLICAN']
            democrat_vote = int(democrat_vote_row['candidatevotes'])
            republican_vote = int(republican_vote_row['candidatevotes'])
            democrat_votes.append(democrat_vote)
            republican_votes.append(republican_vote)
        except Exception:
            democrat_votes.append(0)
            republican_votes.append(0)
    df_to_fill['democrat_votes'] = democrat_votes
    df_to_fill['republican_votes'] = republican_votes
    df_to_fill = df_to_fill.loc[df_to_fill['democrat_votes'] != 0]
    df_to_fill = df_to_fill.loc[df_to_fill['republican_votes'] != 0]
    df_to_fill.to_csv(csv_name, index=False)

fill_vote_for_df('confirmed_deaths_10_9_2021.csv')
```

```python
vote_df = pd.read_csv('countypres_2000-2020_2.csv')
vote_df = vote_df[['county_fips', 'party', 'candidatevotes', 'totalvotes']]

def calculate_political_percentages(csv_name):
    df_to_fill = pd.read_csv(csv_name)
    democrat_percents = []
    republican_percents = []
    for index, row in df_to_fill.iterrows():
        county_fips = row['FIPS']
        county_vote_rows = vote_df.loc[vote_df['county_fips'] == county_fips]
        democrat_vote_row = county_vote_rows.loc[county_vote_rows['party'] == 'DEMOCRAT']
        republican_vote_row = county_vote_rows.loc[county_vote_rows['party'] == 'REPUBLICAN']
        democrat_votes = int(democrat_vote_row['candidatevotes'])
        republican_votes = int(republican_vote_row['candidatevotes'])
        total_vote = int(democrat_vote_row['totalvotes'])
        democrat_percent = (democrat_votes / total_vote) * 100
        republican_percent = (republican_votes / total_vote) * 100
        democrat_percents.append(democrat_percent)
        republican_percents.append(republican_percent)
    df_to_fill['democrat_percent'] = democrat_percents
    df_to_fill['republican_percent'] = republican_percents

    df_to_fill.to_csv(csv_name, index=False)

calculate_political_percentages('confirmed_deaths_10_9_2021.csv')
```

```python
def fill_confirmed_cases(csv_name, date):
    confirmed_cases_df = pd.read_csv('time_series_covid19_confirmed_US.csv')
    confirmed_cases_df = confirmed_cases_df[['FIPS', date]]
    df_to_fill = pd.read_csv(csv_name)
    confirmed_cases_list = []
    for index, row in df_to_fill.iterrows():
        county_fips = row['FIPS']
        confirmed_cases = int(confirmed_cases_df.loc[confirmed_cases_df['FIPS'] == county_fips][date])
        confirmed_cases_list.append(confirmed_cases)
    df_to_fill['confirmed_cases'] = confirmed_cases_list
    df_to_fill.to_csv(csv_name, index=False)

fill_confirmed_cases('confirmed_deaths_10_9_2021.csv', '10/9/21')
```
